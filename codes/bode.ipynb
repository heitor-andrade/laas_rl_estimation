```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_impedance(R, f, L = 200e-6):
    return np.sqrt(R**2 + (2*np.pi*f*L)**2)

R = 0.53
f_init, f_end = 100, 5000
fs = list(np.logspace(np.log10(f_init), np.log10(f_end)))
fs = [0] + fs
# fs = np.logspace(1, 7)

Zs = [compute_impedance(R, f) for f in fs]
# plt.plot(fs, Zs);
# plt.grid()
U = 1

currents_ac = [U/Z for Z in Zs]
current_dc = U/R
gain_currents = [current/current_dc for current in currents_ac]

plt.plot(fs, 20*np.log10(gain_currents) );
plt.xlabel("Frequence [Hz]")
plt.ylabel("Gain Current [Db]")
plt.title("Theoretical current for L = 200e-6 and R = 0.53")
plt.xscale("log")
plt.grid()
plt.legend();
```
